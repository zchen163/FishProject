```python
path = "../data/CorcoranLake_2018pics/"
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
lms = readtps('./trainingset/corcoran_landmarks.tps')
marks = lms["coords"]
for f in range(len(onlyfiles[:10])):
    img = cv2.imread(join(path, onlyfiles[f]))
    fig, ax = plt.subplots()
    ax.imshow(cv2.cvtColor(cv2.flip(img, 0), cv2.COLOR_BGR2RGB), cmap=plt.cm.gray)
    ax.plot(marks[:, 0, f], marks[:, 1, f], '.b', lw=3)
#     for i in range(marks.shape[0]): 
#         ax.plot(marks[i, 0], marks[i, 1], '.b', lw=3)
```
